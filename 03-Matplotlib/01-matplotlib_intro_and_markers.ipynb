```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#   Plot (x-axis, y-axis) to draw a graph

x_points= np.array([1,15])
y_points= np.array([100,250])

plt.plot(x_points, y_points)
plt.show()
```

Plotting Without Line

```python
#   To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

x_points= np.array([1,8])
y_points= np.array([3,10])

plt.plot(x_points, y_points, "o")
plt.show()
```

```python
#   Multiple Points

x_pt= np.array([1,2,3,4])
y_pt= np.array([4,1,5,7])

plt.plot(x_pt, y_pt)
```

```python
#   Plotting without x-points:

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

```python
#   Mark each point with a circle:

y= np.array([4,8,7,1])

#plt.plot(y, marker="o")
#plt.show

plt.plot(y, marker="*")
plt.show
```

Format Strings fmt
You can also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this syntax:

marker|line|color

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

Marker Size

```python
#   You can use the keyword argument markersize or the shorter version, ms to set the size of the markers:

ypts = np.array([3, 8, 1, 10])

plt.plot(ypts, marker = '*', ms = 17)
plt.show()
```

Marker Color

```python
#   You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:

ypts = np.array([3, 8, 1, 10])

plt.plot(ypts, marker = '*', ms = 17, mec="m")
plt.show()
```

```python
#   You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers:

ypts = np.array([3, 8, 1, 10])

plt.plot(ypts, marker = '*', ms = 17, mec="m", mfc="y")
plt.show()

#   Use both the mec and mfc arguments to color the entire marker.
#   You can also use Hexadecimal color values.Or any of the 140 supported color names.(140 supported color names: https://www.w3schools.com/colors/colors_names.asp)
```
